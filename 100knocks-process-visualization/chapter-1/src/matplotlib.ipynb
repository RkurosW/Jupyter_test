# matplotlib

```python
import matplotlib.pyplot as plt
import os
output_dir = '../data/output'
```

```python
fig = plt.figure(figsize=(5,5))
plt.plot([1, 2, 3, 4, 5], [2, 4, 8, 16, 32]) # 折れ線グラフ
plt.plot([1, 2, 3, 4, 5], [1, 4, 9, 16, 25], c='red', alpha=0.5, linewidth=3)
```

```python
plt.figure(figsize=(5, 5))
x = [1, 2, 3, 4, 5]
plt.plot(x, [2, 4, 8, 16, 32], label='y=2^x')
plt.plot(x, [1, 4, 9, 16, 25], '--', label='y=x^2')
plt.plot(x, [2, 8, 10, 12, 20], '-.', label='y=2x^2')
# plt.plot(x, [4, 10, 16, 22, ], '-.', label='y=4x^2')
plt.legend()
```

```python
plt.figure(figsize=(5, 5))
x = [1, 2, 3, 4, 5]
y = [30, 90, 60, 20, 70]
label = ['A', 'B', 'C', 'D', 'E']
plt.bar(x, y, tick_label=label) # 棒グラフ
```

```python
plt.figure(figsize=(3, 3))
plt.scatter([1, 2, 3, 4, 5], [2, 4, 8, 16, 32], label='y=2^x', marker='o') # 散布図
```

```python
import numpy as np
x = np.random.rand(30)
y = np.random.rand(30)
plt.figure(figsize=(5, 5))
plt.scatter(x, y, s=50, c='blue') # 散布図
```

```python
label = ['A', 'B', 'C', 'D', 'E', 'F']
x = [np.random.randint(1, 30) for _ in range(len(label))]
print(x)

plt.figure(figsize=(3, 3))
plt.pie(x, labels=label, counterclock=False, startangle=90)
```

```python
fig, ax = plt.subplot(2,2, igsize=(3, 3))
ax[0][0].plot([1, 2, 3, 4], [1, 4, 9, 16])
ax[0][0].set_title("y=x^2")

ax[1][1].pie([20, 30, 35, 15], labels=['A', 'B', 'C', 'D'])
```
